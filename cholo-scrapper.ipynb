# First attempt to make an AI Web Scrapper 

The twist here is that is not only a web scrapper, it's going to analyze the content of the website, and then summarize it like a cholo (you hear that right foo). Also we are going to be using Ollama. The model for now as this first commit is llama3.2, because i'm using a 2020 Macbook Air (lol)


```python
import ollama 
from IPython.display import Markdown, display
from scrapper import fetch_website_posts 
```

So the first thing is create a system and a user prompt, in here, is important to note two things. And this is for myself, it's super basic that almost everybody know this, but it's just in case i need to refresh my memory

    * System prompt: is the context (Overall information about the task)
    * User prompt: is the message (End user give in the context given)
  

```python
system_prompt = """
You are a snarky cholo that went to college, your favorite things amongst others are analyze the data from websites,  
hanging out with the homies, do some graffiti, bump some lowriders and provide a short summaries of what you read 
you are very professional and you like to ignore text that might be irrelevant or navigation related. 
Respond in markdown format. And in spanish, but like the cholo you are
"""

user_prompt = """
Here is a website, provide a short summary of the data present in here. 
"""
```

Our next step is to create a basic web scrapper, so here we are going to reference our file ==scrapper.py== and then create a new function called summarize

```python
def summarize(url: str):
    content = fetch_website_posts(url)       
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + "\n\n" + content}
    ]
    response = ollama.chat(model="llama3.2:3b", messages=messages)
    return response.message.content
```

We are almost there, now we need a function that shows the work of our lil' homie

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

Lastly we present the information

```python
display_summary("")
```
